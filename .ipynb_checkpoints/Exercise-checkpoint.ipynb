```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import KMeans
```

```python
df=pd.read_csv('country_data.csv')
df.head()
```

```python
sns.pairplot(df)
```

```python
X=df.drop(['country'],axis=1)
X
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scalar=StandardScaler()
sc=scalar.fit_transform(X)
features=scalar.get_feature_names_out()
```

```python
features
```

```python
x_scaled=pd.DataFrame(sc,columns=features)
x_scaled
```

```python
sse=[]
krng=range(4,15)
for k in krng:
    km=KMeans(n_clusters=k, n_init='auto')
    km.fit(x_scaled)
    sse.append(km.inertia_)
sse
```

```python
plt.plot(krng,sse)
```

```python
km2=KMeans(n_clusters=6, n_init='auto')
km2.fit(x_scaled)
```

```python
df['cluster']=km2.labels_
df
```

```python
sns.pairplot(df,hue='cluster',palette='viridis')
# sns.pairplot(x=km2.cluster_centers_[:,0],y=km.cluster_centers_[:,1],marker='*',s=200,colors='orange')
```

```python
km2.cluster_centers_
```

```python
hc = AgglomerativeClustering(n_clusters=6, metric='cosine', linkage='complete')
hc.fit(x_scaled)

#Assign cluster labels to the original DataFrame and print a sample of 10 rows to verify the cluster assignments.
df['Cluster_HC'] = hc.labels_
df.head()
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
```

```python
z=linkage(x_scaled, method='complete')
```

```python
plt.figure(figsize=(20, 10))
plt.title('Hierarchical Clustering Dendrogram (Subset)')
dendrogram(z, labels=df['country'].values, leaf_rotation=90, leaf_font_size=10, above_threshold_color='black')
plt.xlabel('Country')
plt.ylabel('Distance')
plt.grid(True)
plt.xticks(rotation=90, fontsize=10)
plt.tight_layout()
plt.show()
```
